# Mathematical Background on Legendre Transformation

In the following section we try to give a more geometric perspective on what the Legendre transformation is and how it can be used to transform between different but equivalent descriptions of a mechanical system. From a mathematica point of view, the Legendre transformation is a way to convert a function defined on a vector space into a function defined on the dual vector space (which is just the space of linear functionals on the original vector space). 

For simplicity, we consider first the case of a single variable function f defined on the real line (which is a one-dimensional vector space). Thus let f: R -> R be a smooth and convex function. In order to have some picture in mind we can take as an example the function $f(x) = e^x$ defined for all real numbers x. 

The graph of the function f is a curve in the plane and we can obviously represent such function as a set of points (x, f(x)) in the plane. This is usual way how we think about functions and their graphs. 

However, there is another way to represent the same function which is more geometric and which will be useful for our purposes. We can represent the function f as a set of tangent lines to the graph of f. Since the function f is convex, at each point x there is a unique tangent line to the graph of f with an unique slope. Of course, the tangent line is always unique, but convexity guarantees also that there is one to one mapping between the points on the graph and the tangent lines.



````python
# In order to illustrate the above point we plot the function f and its tangent lines in an 
# intervals around the point x = 0. We can see that the tangent lines are in one to one correspondence with the points on the graph of f.
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2, 2, 100)
f = np.exp(x)
plt.plot(x, f, label='f(x) = exp(x)', color = 'red', linewidth = 4)
for x0 in np.linspace(-1.5, 1.5, 20):
    slope = np.exp(x0)
    intercept = np.exp(x0) - slope * x0
    tangent_line = slope * x + intercept
    plt.plot(x, tangent_line, label=f'Tangent at x={x0:.1f}', color = 'blue', alpha = 0.5)
plt.title('Duality between points on the graph and tangent lines')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(-0.25, 5.0)
plt.grid()
plt.show()
````

The illustration above illustrates the duality between the graph of the function $f$ as a set of points (x, f(x)) and the graph of the function $f^*$ as a set of tangent lines to the graph of f. 

Now we can define a linear functional which is an element of the dual vector space. In our case, since we are working with a one-dimensional vector space, the dual vector space is also one-dimensional and we can represent its elements as real numbers. Any linear functional on the real line can be represented as a real number p which acts on a point x by multiplication: $px = p \cdot x$. With the help of such functional we can define the following function 

$$
F(x,p) = px - f(x)
$$

For a given value of p, the function F(x,p) is a function of x and we can find its maximum value with respect to x. This defines a new function, which then only depends on p, and is given by

$$
g(p) = \max_x F(x,p) = \max_x (px - f(x))
$$

What does this mean geometrically?

For fixed p, $F(x,p)=px-f(x)$ is the signed vertical difference between f(x) and the line $y=px$. Maximizing it over $x$ delivers the point where the tangent line to f has exactly the slope p. Then, $g(p)$ is exactly the intercept of the corresponding tangent line given by $y=px-g(p)$. In this way, we get a new representation of the function f, in terms of pairs of valuses $(p, g(p))$ which correspond to the slope and the intercept of the tangent lines to the graph of f.

## Legendre transformation is an involution 

To Legendre-transform $g(p)$ once more, we define

$$
h(x) = \max_p (xp - g(p)).
$$

At the maximum, the derivative with respect to $p$ must vanish, so

$$
0 = \frac{\partial}{\partial p}(xp-g(p)) = x - \frac{dg}{dp}.
$$

Hence

$$
x = \frac{dg}{dp}.
$$

Now recall that

$$
g(p) = p x(p) - f(x(p)),
$$

where $p = f'(x(p))$. Differentiating gives

$$
\frac{dg}{dp} = x(p) + \big(p - f'(x(p))\big)\frac{dx}{dp} = x(p).
$$

Therefore the maximizing point satisfies $x = x(p)$, or equivalently $p = p(x)$. Substituting this into $h(x)$ yields

$$
h(x) = xp(x) - g(p(x)).
$$

Using again the definition of the Legendre transform,

$$
g(p(x)) = p(x)x - f(x),
$$

we obtain

$$
h(x) = xp(x) - \big(p(x)x - f(x)\big) = f(x).
$$

So applying the Legendre transformation twice returns the original function. This is why the Legendre transformation is called an involution.
