## Position and Momentum Representation

### Position as an Quantum-mechanical Observable

We consider a quantum system whose configuration space is the real line. The position of the particle is an observable that can take any real value not just discrete ones. According to the rules of quantum mechanics, an observable is represented by a self-adjoint operator. We therefore associate to position an operator $\hat x$ satisfying

$$
\hat x^\dagger=\hat x .
$$

Let $|\psi\rangle$ be a normalized state in the Hilbert space $\mathcal H$:

$$
\langle \psi|\psi\rangle =1 .
$$

Using rules of quantum mechanics we can associate the  expectation value of position to the state $|\psi\rangle$ according to

$$
\langle \hat x\rangle_\psi
=
\langle \psi|\hat x|\psi\rangle .
$$

This gives us a real number that represents the mean position that would be obtained by repeating position measurements on identically prepared systems, all in state $|\psi\rangle$. One should notice the expectation value $\langle \hat x\rangle_\psi$ does **not** uniquely label the state. Many different quantum states can have the same average position. 

## Translation Operator and Its Generator

We have seen in classical mechanics that the momentum as a classical observable can be understood as the infinitesimal generator of translations in space. Following this approach we introduce the momentum operator in quantum mechanics by **defining** it as the generator of spatial translations. 

Intuitively, the translation moves the quantum state in space by a certain amount $\Delta x$ but it does not change the overall "shape" of the wavefunction nor its normalization. We can think of the translation as a symmetry transformation that leaves the physics invariant but changes the position of the particle. Motivated by this, we now define a unitary operator called the **translation operator**, denoted by $\hat{T}(\Delta x)$, that shifts a quantum state in space by a distance $\Delta x$. Specifically, this operator satisfies:

$$
\hat{T}^\dagger(\Delta x) \hat{T}(\Delta x) = \hat{T}(\Delta x) \hat{T}^\dagger(\Delta x) = \mathbb{I},
$$

i.e., $\hat{T}(\Delta x)$ is **unitary** for all $\Delta x \in \mathbb{R}$. $\Delta x$ itself is a real parameter that labels each translation operator. 



The tranlation operators form a continuous group under the operation of composition, known as the **translation group**. The composition of two translations corresponds to the addition of their parameters: 

$$\hat{T}(\Delta x_1) \hat{T}(\Delta x_2) = \hat{T}(\Delta x_1 + \Delta x_2).$$

The identity element of this group is the operator $\hat{T}(0) = \mathbb{I}$, which corresponds to no translation. The inverse of a translation operator is given by $\hat{T}(-\Delta x)$, which undoes the translation by $\Delta x$.



### Shift of Expectation Value

To understand the action of the translation operator, we consider a state $|\psi_{\langle x \rangle}\rangle$ that has a well-defined expectation value of position $\langle x \rangle$. We want to see how the translation operator shifts this expectation value.
To fix the form of $\hat{T}(\Delta x)$, we demand that it shifts the expectation value of position for **every** state:

$$
\left\langle \psi_{\langle x \rangle} \right| \hat{T}^\dagger(\Delta x) \hat{x} \hat{T}(\Delta x) \left| \psi_{\langle x \rangle} \right\rangle = \langle x \rangle + \Delta x.
$$

This condition implies the operator identity:

$$
\hat{T}^\dagger(\Delta x) \hat{x} \hat{T}(\Delta x) = \hat{x} + \Delta x \cdot \mathbb{I}.
$$

Multiplying both sides from the left with $\hat{T}(\Delta x)$, we obtain the **commutator relation**:

$$
\hat{x} \hat{T}(\Delta x) - \hat{T}(\Delta x) \hat{x} = \Delta x \cdot \hat{T}(\Delta x),
$$

or

$$
[\hat{x}, \hat{T}(\Delta x)] = \Delta x \cdot \hat{T}(\Delta x).
$$

We observe that the translation operator does not commute with the position operator!




## Momentum as Generator of Translations

To explore the structure of the translation operator, we differentiate the unitarity condition with respect to the translation parameter that we now label by $x$ (we now write $\hat{T}(x)$ instead of $\hat{T}(\Delta x)$ to emphasize that $x$ is a continuous parameter):

$$
\hat{T}^\dagger(x) \hat{T}(x) = \mathbb{I}.
$$

Formally differentiating both sides with respect to $x$, we obtain

$$
\frac{\partial \hat{T}}{\partial x} \hat{T}^\dagger + \hat{T} \frac{\partial \hat{T}^\dagger}{\partial x} = 0.
$$



This implies that

$$
\left( 
\frac{\partial \hat T}{\partial x}\hat T^\dagger
\right)
+
\left(
\frac{\partial \hat T}{\partial x}\hat T^\dagger
\right)^\dagger
=0,
$$

which shows that the operator

$$
\frac{\partial \hat T}{\partial x}\hat T^\dagger
$$

is anti-Hermitian. Notice the exact analogy with the case of time evolution, where we found that the operator $\frac{\partial \hat U}{\partial t}\hat U^\dagger$ is anti-Hermitian.

Thus, to every continuously parametrized unitary operator $\hat T(x)$ we can associate an anti-Hermitian operator

$$
\pi(x)
=
\frac{\partial \hat T(x)}{\partial x}\hat T^\dagger(x),
$$

with

$$
\pi^\dagger(x)=-\pi(x).
$$

At this point, $\pi(x)$ may still depend on $x$. For spatial translations, however, we require more than unitarity! 
Translations must satisfy the group property

$$
\hat T(x+y)=\hat T(x)\hat T(y),
$$

together with

$$
\hat T(0)=\mathbb I.
$$

This expresses the physical fact that translating first by $y$ and then by $x$ is equivalent to translating once by $x+y$.

Now consider the derivative of $\hat T(x)$ as it is defined by the limit (strong limit in the Hilbert space):

$$
\frac{\partial \hat T(x)}{\partial x}
=
\lim_{\epsilon\to 0}
\frac{\hat T(x+\epsilon)-\hat T(x)}{\epsilon}.
$$

Using the group property,

$$
\hat T(x+\epsilon)=\hat T(\epsilon)\hat T(x),
$$

we obtain

$$
\frac{\partial \hat T(x)}{\partial x}
=
\lim_{\epsilon\to 0}
\frac{\hat T(\epsilon)-\mathbb I}{\epsilon}
\hat T(x).
$$

Now we **define** the infinitesimal generator as

$$
\pi
=
\left.
\frac{\partial \hat T(\epsilon)}{\partial \epsilon}
\right|_{\epsilon=0}.
$$

and observe that it is independent of $x$! Therefore,

$$
\frac{\partial \hat T(x)}{\partial x}
=
\pi \hat T(x).
$$

Multiplying from the right by $\hat T^\dagger(x)$ gives

$$
\frac{\partial \hat T(x)}{\partial x}\hat T^\dagger(x)
=
\pi .
$$

The operator differential equation

$$
\frac{\partial \hat T(x)}{\partial x}
=
\pi \hat T(x)
$$

with initial condition

$$
\hat T(0)=\mathbb I
$$

has the formal solution

$$
\hat T(x)=e^{\pi x}.
$$



Since $\pi$ is anti-Hermitian, we can write it as $-i$ times a Hermitian operator. We therefore **define the momentum operator** $\hat p$ by the identity

$$
\pi=-\frac{i}{\hbar}\hat p.
$$

Equivalently,

$$
\hat p=i\hbar \pi.
$$

Because $\pi^\dagger=-\pi$, it follows that

$$
\hat p^\dagger=\hat p.
$$

Thus $\hat p$ is Hermitian, as required for an observable.

The translation operator becomes

$$
\hat T(x)
=
e^{-\frac{i}{\hbar}\hat p x}.
$$

For some finite displacement $\Delta x$, we can therefore write

$$
\hat T(\Delta x)
=
e^{-\frac{i}{\hbar}\hat p \Delta x}.
$$

Hence, the momentum operator in quantum mechanics is the Hermitian generator of spatial translations.


### Some more math background: Translation as a One-Parameter Unitary Group

From a mathematical point of view, the family of translation operators $\{ \hat{T}(x) \}_{x \in \mathbb{R}}$ forms a **one-parameter unitary group** under composition as a group operation:

1. $\hat{T}(0) = \mathbb{I}$ (identity element),
2. $\hat{T}(x_1)\hat{T}(x_2) = \hat{T}(x_1 + x_2)$ (group property),
3. $\hat{T}(x)$ is continuous in the parameter $x \in \mathbb{R}$.

These properties imply that translations in space form a continuous group structure where each group element is labeled by a real parameter $x$, and multiplication of translation operators corresponds to addition of their displacement parameters.

### Connection to Stone’s Theorem

According to **Stone’s theorem** on one-parameter unitary groups, every (strongly) continuous one-parameter unitary group $\{ U(t) \}_{t \in \mathbb{R}}$ on a Hilbert space $\mathcal{H}$ can be written in the form

$$
U(t) = e^{-i t A}
$$

for some unique **self-adjoint operator** $A$, which is called the **generator** of the unitary group.

In our case, the group $\hat{T}(x)$ is parametrized by the real displacement $x$, and the generator of spatial translations is a Hermitian operator which we identify with the **momentum operator**:

$$
\hat{T}(x) = e^{- \frac{i}{\hbar} \hat{p} x}.
$$

This shows that the momentum operator $\hat{p}$ is (up to a factor $\hbar$) the generator of spatial translations in quantum mechanics meaning that the infinitesimal translations are induced by the action of $\hat{p}$ on a given state.


## Commutator of $\hat{x}$ and $\hat{p}$

We now ask whether position and momentum can be simultaneously sharp. In operator language, this means asking whether the position and momentum operators commute.

From the transformation property of the position operator under translations, we previously obtained

$$
[\hat x,\hat T(\Delta x)] = \Delta x\,\hat T(\Delta x).
$$

Using the exponential form of the translation operator,

$$
\hat T(\Delta x)=e^{-\frac{i}{\hbar}\hat p\,\Delta x},
$$

we get

$$
\left[
\hat x,
e^{-\frac{i}{\hbar}\hat p\,\Delta x}
\right]
=
\Delta x\, e^{-\frac{i}{\hbar}\hat p\,\Delta x}.
$$

We now differentiate both sides with respect to the translation parameter $\Delta x$:

$$
\left[
\hat x,
-\frac{i}{\hbar}\hat p\,
e^{-\frac{i}{\hbar}\hat p\,\Delta x}
\right]
=
e^{-\frac{i}{\hbar}\hat p\,\Delta x}
+
\Delta x
\left(
-\frac{i}{\hbar}\hat p
\right)
e^{-\frac{i}{\hbar}\hat p\,\Delta x}.
$$

Now set $\Delta x=0$. Since

$$
e^0=\mathbb I,
$$

we obtain

$$
\left[
\hat x,
-\frac{i}{\hbar}\hat p
\right]
=
\mathbb I .
$$

Therefore,

$$
-\frac{i}{\hbar}[\hat x,\hat p]=\mathbb I,
$$

and hence

$$
[\hat x,\hat p]=i\hbar\mathbb I.
$$

Thus, position and momentum do not commute. Their canonical commutation relation is

$$
[\hat x,\hat p]=i\hbar\mathbb I.
$$

This non-commutativity is the algebraic origin of the Heisenberg uncertainty relation.


## On the Eigenstates of Position and Momentum

The fact that $\hat{x}$ and $\hat{p}$ do not commute implies that there is **no basis** in which both operators are diagonal—there are **no common eigenstates**.

Assume $\hat{x}$ has eigenstates $|x\rangle$ satisfying:

$$
\hat{x} |x\rangle = x \cdot |x\rangle.
$$

We assume these states form a **complete set** in the Hilbert space:

$$
\int dx\, |x\rangle\langle x| = \mathbb{I}.
$$

Then any state $|\psi\rangle$ can be decomposed as:

$$
|\psi\rangle = \int dx' \, |x'\rangle \langle x' | \psi \rangle.
$$

Define the **position-space wavefunction** as

$$
\psi(x') := \langle x' | \psi \rangle.
$$

Then projecting onto $\langle x|$ gives:

$$
\langle x | \psi \rangle = \int dx' \langle x | x' \rangle \psi(x').
$$

This shows that $\langle x | x' \rangle$ acts like a **Dirac delta function**:

$$
\langle x | x' \rangle = \delta(x - x'), \quad \text{with} \quad \int dx' \delta(x - x') = 1.
$$



## Action of Operators in Position Representation

Let us now examine how operators act in the position representation.

### Position operator

We compute:

$$
\langle x | \hat{x} | \psi \rangle = x \cdot \langle x | \psi \rangle = x \cdot \psi(x).
$$

Thus:

$$
\hat{x} \psi(x) = x \cdot \psi(x).
$$

This means that in the position representation, the position operator acts as **multiplication by $x$**.




### Momentum operator

We now compute how $\hat{p}$ acts on a wavefunction $\psi(x)$. Recall that

$$
\hat{T}(\Delta x) = e^{- \frac{i}{\hbar} \hat{p} \Delta x}
\approx \mathbb{I} - \frac{i}{\hbar} \hat{p} \Delta x + \mathcal{O}(\Delta x^2),
$$

so that:

$$
\hat{p} = i \hbar \lim_{\Delta x \to 0} \frac{\hat{T}(\Delta x) - \mathbb{I}}{\Delta x}.
$$

Then:

$$
\begin{aligned}
\langle x | \hat{p} | \psi \rangle
&= i \hbar \lim_{\Delta x \to 0} \frac{1}{\Delta x} \left( \langle x | \hat{T}(\Delta x) | \psi \rangle - \langle x | \psi \rangle \right) \\
&= i \hbar \lim_{\Delta x \to 0} \frac{1}{\Delta x} \left( \psi(x - \Delta x) - \psi(x) \right) \\
&= -i \hbar \frac{d\psi}{dx}(x).
\end{aligned}
$$

Hence:

$$
\hat{p} \psi(x) = -i \hbar \frac{d}{dx} \psi(x).
$$

So, in the position representation, the momentum operator acts as a **differential operator**:

$$
\hat{p} = -i \hbar \frac{d}{dx}.
$$




## Heisenberg’s Uncertainty Relation


We now examine to what extent the position and the momentum of a quantum particle can be simultaneously determined. We define the **uncertainty** in the position and in the momentum as the expectation value of the squared deviation from the mean:

$$
\begin{aligned}
& \left\langle(\Delta x)^2\right\rangle = \langle \psi | (\hat{x} - \langle x \rangle)^2 | \psi \rangle \\
& \left\langle(\Delta p)^2\right\rangle = \langle \psi | (\hat{p} - \langle p \rangle)^2 | \psi \rangle
\end{aligned}
$$

As a measure of the uncertainty we use the product of the position and momentum variances:

$$
\left\langle(\Delta x)^2\right\rangle \cdot \left\langle(\Delta p)^2\right\rangle = 
\langle \psi | (\hat{x} - \langle x \rangle)^2 | \psi \rangle \cdot \langle \psi | (\hat{p} - \langle p \rangle)^2 | \psi \rangle.
$$

Since both $ \hat{x} - \langle x \rangle $ and $ \hat{p} - \langle p \rangle $ are self-adjoint operators, we can write these variances as squared norms of the corresponding vectors:

$$
\begin{aligned}
\left\langle(\Delta x)^2\right\rangle &= \| (\hat{x} - \langle x \rangle) \psi \|^2 \\
\left\langle(\Delta p)^2\right\rangle &= \| (\hat{p} - \langle p \rangle) \psi \|^2
\end{aligned}
$$



Their product is therefore:

$$
\left\langle(\Delta x)^2\right\rangle \cdot \left\langle(\Delta p)^2\right\rangle 
= \| (\hat{x} - \langle x \rangle) \psi \|^2 \cdot \| (\hat{p} - \langle p \rangle) \psi \|^2.
$$

We now apply the **Cauchy-Schwarz inequality**:

$$
| \langle a | b \rangle |^2 \leq \|a\|^2 \cdot \|b\|^2,
$$

with $a = (\hat{x} - \langle x \rangle)\psi$ and $b = (\hat{p} - \langle p \rangle)\psi$. This yields:

$$
\left\langle(\Delta x)^2\right\rangle \cdot \left\langle(\Delta p)^2\right\rangle 
\geq \left| \langle (\hat{x} - \langle x \rangle)\psi \mid (\hat{p} - \langle p \rangle)\psi \rangle \right|^2.
$$

Let us denote this inner product as:

$$
\langle (\Delta x)\psi \mid (\Delta p)\psi \rangle = \langle \psi | (\hat{x} - \langle x \rangle)(\hat{p} - \langle p \rangle) | \psi \rangle.
$$


We can now define the fluctuation operators as:

$$
\Delta \hat{x} = \hat{x}-\langle \hat{x}\rangle,
\qquad
\Delta \hat{p} = \hat{p}-\langle \hat{p}\rangle .
$$

From the Cauchy--Schwarz inequality applied to the two states $\Delta \hat{x}|\psi\rangle$ and $\Delta \hat{p}|\psi\rangle$, we obtain

$$
\langle (\Delta \hat{x})^2\rangle
\langle (\Delta \hat{p})^2\rangle
\geq
\left|\langle \Delta \hat{x}\Delta \hat{p}\rangle\right|^2 .
$$

Now introduce

$$
A =
\langle \Delta \hat{x}\Delta \hat{p}\rangle .
$$

This quantity is in general complex. Therefore,

$$
|A|^2
=
(\operatorname{Re}A)^2
+
(\operatorname{Im}A)^2
\geq
(\operatorname{Im}A)^2 .
$$

The imaginary part of $A$ is determined by the commutator:

$$
\begin{aligned}
\operatorname{Im}A
&=
\frac{1}{2i}
\left(A-A^*\right) \\[4pt]
&=
\frac{1}{2i}
\left(
\langle \Delta \hat{x}\Delta \hat{p}\rangle
-
\langle \Delta \hat{p}\Delta \hat{x}\rangle
\right) \\[4pt]
&=
\frac{1}{2i}
\langle [\Delta \hat{x},\Delta \hat{p}]\rangle .
\end{aligned}
$$

Since constants commute with all operators, we have

$$
[\Delta \hat{x},\Delta \hat{p}]
=
[\hat{x},\hat{p}]
=
i\hbar .
$$

Hence,

$$
\operatorname{Im}A
=
\frac{1}{2i}
\langle i\hbar\rangle
=
\frac{\hbar}{2}.
$$

It follows that

$$
\left|\langle \Delta \hat{x}\Delta \hat{p}\rangle\right|^2
\geq
\left(\frac{\hbar}{2}\right)^2 .
$$

Combining this with the Cauchy--Schwarz inequality gives

$$
\langle (\Delta \hat{x})^2\rangle
\langle (\Delta \hat{p})^2\rangle
\geq
\left(\frac{\hbar}{2}\right)^2 .
$$

Taking the square root yields

$$
\Delta x\,\Delta p
\geq
\frac{\hbar}{2},
$$

where

$$
\Delta x =
\sqrt{\langle(\Delta \hat{x})^2\rangle},
\qquad
\Delta p =
\sqrt{\langle(\Delta \hat{p})^2\rangle}.
$$

This is the celebrated position--momentum uncertainty relation. Historically, Werner Heisenberg introduced it in 1927 from considerations of measurement (Gedankenexperimente!). The mathematically precise inequality derived here is an expression of the fundamental limit on the simultaneous knowledge of position and momentum in quantum mechanics, and it a direct consequence of the non-commutativity of the position and momentum operators.    

### Minimum uncertainty wave packet and derivation of the Fourier kernel

We have seen that there is a lower bound to the product of the uncertainties in position and momentum,

$$
\sigma_x \sigma_p \geq \frac{\hbar}{2}.
$$

We now ask whether this lower bound can actually be reached, i.e. whether there are states for which

$$
\sigma_x \sigma_p = \frac{\hbar}{2}.
$$

Such states are called **minimum uncertainty states**.

We define the shifted operators

$$
X = \hat{x} - \langle x \rangle,
$$

and

$$
P = \hat{p} - \langle p \rangle.
$$

The equality condition in the Cauchy--Schwarz inequality implies that the two vectors $X|\psi\rangle$ and $P|\psi\rangle$ must be linearly dependent. Hence

$$
P|\psi\rangle = \lambda X|\psi\rangle,
$$

or explicitly,

$$
(\hat{p} - \langle p \rangle)|\psi\rangle
=
\lambda(\hat{x} - \langle x \rangle)|\psi\rangle,
$$

where $\lambda \in \mathbb{C}$.

Normalizability will require that the imaginary part of $\lambda$ is positive. For simplicity, we choose $\lambda$ to be purely imaginary,

$$
\lambda = ia,
$$

with

$$
a>0.
$$



#### Solution in position representation

We project the minimum uncertainty condition onto position eigenstates. In position representation,

$$
\hat{x} = x,
$$

and

$$
\hat{p} = -i\hbar \frac{d}{dx}.
$$

Therefore

$$
\left(
-i\hbar \frac{d}{dx} - \langle p \rangle
\right)\psi(x)
=
\lambda
\left(
x - \langle x \rangle
\right)\psi(x).
$$

This is a first-order differential equation. Rearranging gives

$$
\frac{d\psi}{dx}
=
\frac{i}{\hbar}
\left[
\lambda(x-\langle x\rangle)
+
\langle p \rangle
\right]\psi(x).
$$

Integrating directly, we obtain

$$
\psi(x)
=
A
\exp
\left[
\frac{i\lambda}{2\hbar}(x-\langle x\rangle)^2
+
\frac{i\langle p\rangle}{\hbar}x
\right].
$$

Now we choose

$$
\lambda = ia.
$$

Then



$$
\psi(x)
=
A
\exp
\left[
-\frac{a}{2\hbar}(x-\langle x\rangle)^2
+
\frac{i\langle p\rangle}{\hbar}x
\right].
$$

We introduce the width $\sigma_x$ by

$$
a = \frac{\hbar}{2\sigma_x^2}.
$$

Thus the normalized position-space wave function is

$$
\psi(x)
=
N_x
\exp
\left[
-\frac{(x-\langle x\rangle)^2}{4\sigma_x^2}
+
\frac{i\langle p\rangle}{\hbar}x
\right],
$$

where

$$
N_x
=
\left(
\frac{1}{2\pi\sigma_x^2}
\right)^{1/4}.
$$



#### Solution in momentum representation

We now solve the same equation in momentum representation. In momentum space,

$$
\hat{p} = p,
$$

and

$$
\hat{x} = i\hbar \frac{d}{dp}.
$$

Therefore the minimum uncertainty condition becomes

$$
(p-\langle p\rangle)\phi(p)
=
\lambda
\left(
i\hbar\frac{d}{dp}
-
\langle x\rangle
\right)\phi(p).
$$

Equivalently,

$$
(p-\langle p\rangle)\phi(p)
=
i\hbar\lambda\frac{d\phi}{dp}
-
\lambda\langle x\rangle\phi(p).
$$

Using

$$
\lambda = ia,
$$

we get

$$
\frac{d\phi}{dp}
=
-\frac{p-\langle p\rangle}{\hbar a}\phi(p)
-
\frac{i\langle x\rangle}{\hbar}\phi(p).
$$

Integrating gives

$$
\phi(p)
=
A'
\exp
\left[
-\frac{(p-\langle p\rangle)^2}{2\hbar a}
-
\frac{i\langle x\rangle}{\hbar}p
\right].
$$

Since

$$
a = \frac{\hbar}{2\sigma_x^2},
$$

we have

$$
\frac{1}{2\hbar a}
=
\frac{\sigma_x^2}{\hbar^2}.
$$

For a minimum uncertainty state,

$$
\sigma_x\sigma_p = \frac{\hbar}{2}.
$$

Therefore

$$
\sigma_p = \frac{\hbar}{2\sigma_x}.
$$

Thus

$$
\frac{\sigma_x^2}{\hbar^2}
=
\frac{1}{4\sigma_p^2}.
$$

The normalized momentum-space wave function is therefore

$$
\phi(p)
=
N_p
\exp
\left[
-\frac{(p-\langle p\rangle)^2}{4\sigma_p^2}
-
\frac{i\langle x\rangle}{\hbar}p
\right],
$$

where

$$
N_p
=
\left(
\frac{1}{2\pi\sigma_p^2}
\right)^{1/4}.
$$





#### Derivation of the Fourier kernel

The position and momentum wave functions are two representations of the same abstract state:

$$
\psi(x)=\langle x|\psi\rangle,
$$

and

$$
\phi(p)=\langle p|\psi\rangle.
$$

We insert the momentum-space resolution of the identity,

$$
\hat{1}
=
\int_{-\infty}^{+\infty}
|p\rangle\langle p|\,dp.
$$

Then

$$
\psi(x)
=
\langle x|\psi\rangle
=
\int_{-\infty}^{+\infty}
\langle x|p\rangle
\langle p|\psi\rangle\,dp.
$$

Hence

$$
\psi(x)
=
\int_{-\infty}^{+\infty}
K(x,p)\phi(p)\,dp,
$$

with

$$
K(x,p)=\langle x|p\rangle.
$$



We can now determine the form of $K(x,p)$ by requiring that it transforms the minimum uncertainty wave packet in momentum space into the minimum uncertainty wave packet in position space. Substituting the known forms of $\psi(x)$ and $\phi(p)$ into the integral representation of $\psi(x)$, we obtain:

$$
\begin{aligned}
\psi(x)
&=
N_p
\int_{-\infty}^{+\infty}
K(x,p)
\exp
\left[-\frac{(p-\langle p\rangle)^2}{4\sigma_p^2}
-\frac{i\langle x\rangle}{\hbar}p
\right]dp \\
&=
N_p
\int_{-\infty}^{+\infty}
K(x,p)
\exp
\left[-\frac{(p-\langle p\rangle)^2}{4\sigma_p^2}
+\frac{i}{\hbar}(x-\langle x\rangle)p
\right]dp.
\end{aligned}
$$





We now recall the general formula for Gaussian integrals:   

$$
\int_{-\infty}^{+\infty}
\exp
\left[-\alpha p^2 + \beta p + \gamma
\right]dp
=
\sqrt{\frac{\pi}{\alpha}}
\exp
\left[\frac{\beta^2}{4\alpha} + \gamma
\right],
$$



We now insert the explicit minimum uncertainty wave packets on both sides of

$$
\psi(x)
=
\int_{-\infty}^{+\infty}
K(x,p)\phi(p)\,dp.
$$

The position-space wave packet is

$$
\psi(x)
=
N_x
\exp
\left[
-\frac{(x-\langle x\rangle)^2}{4\sigma_x^2}
+
\frac{i}{\hbar}\langle p\rangle x
\right].
$$

The momentum-space wave packet is

$$
\phi(p)
=
N_p
\exp
\left[
-\frac{(p-\langle p\rangle)^2}{4\sigma_p^2}
-
\frac{i}{\hbar}\langle x\rangle p
\right].
$$

We now ask what the kernel must contribute so that the integral over $p$ produces the known Gaussian in $x$.

The relevant Gaussian integral is

$$
\int_{-\infty}^{+\infty}
\exp
\left[
-\frac{(p-\langle p\rangle)^2}{4\sigma_p^2}
+
Bp
\right]dp
=
\sqrt{4\pi\sigma_p^2}
\exp
\left[
B\langle p\rangle+\sigma_p^2B^2
\right].
$$

The desired position-space Gaussian contains

$$
\exp
\left[
-\frac{(x-\langle x\rangle)^2}{4\sigma_x^2}
\right].
$$

Using

$$
\sigma_x\sigma_p=\frac{\hbar}{2},
$$

we can write this as

$$
\exp
\left[
-\frac{\sigma_p^2}{\hbar^2}(x-\langle x\rangle)^2
\right].
$$

This factor must arise from the term

$$
\exp\left[\sigma_p^2B^2\right].
$$

Therefore we need

$$
B
=
\frac{i}{\hbar}(x-\langle x\rangle).
$$

But the momentum-space wave function already contributes the term

$$
-\frac{i}{\hbar}\langle x\rangle p.
$$

Hence the kernel must contribute the remaining term

$$
\frac{i}{\hbar}xp.
$$

Therefore the kernel must have the form

$$
K(x,p)
=
\mathcal{N}
\exp
\left(
\frac{i}{\hbar}xp
\right).
$$

Substitution gives

$$
\psi(x)
=
\mathcal{N}N_p
\int_{-\infty}^{+\infty}
\exp
\left[
-\frac{(p-\langle p\rangle)^2}{4\sigma_p^2}
+
\frac{i}{\hbar}(x-\langle x\rangle)p
\right]dp.
$$

Using the Gaussian integral with

$$
B
=
\frac{i}{\hbar}(x-\langle x\rangle),
$$

we obtain

$$
\psi(x)
=
\mathcal{N}N_p
\sqrt{4\pi\sigma_p^2}
\exp
\left[
\frac{i}{\hbar}\langle p\rangle(x-\langle x\rangle)
-
\frac{\sigma_p^2}{\hbar^2}(x-\langle x\rangle)^2
\right].
$$

With

$$
\sigma_x\sigma_p=\frac{\hbar}{2},
$$

this becomes

$$
\psi(x)
=
\mathcal{N}N_p
\sqrt{4\pi\sigma_p^2}
\exp
\left[
-\frac{(x-\langle x\rangle)^2}{4\sigma_x^2}
+
\frac{i}{\hbar}\langle p\rangle(x-\langle x\rangle)
\right].
$$

The difference between

$$
\frac{i}{\hbar}\langle p\rangle(x-\langle x\rangle)
$$

and

$$
\frac{i}{\hbar}\langle p\rangle x
$$

is only an $x$-independent phase. Such a phase can be absorbed into the overall normalization constant. Therefore the $x$-dependent part agrees with the known position-space wave packet.

It remains to determine $\mathcal{N}$. Comparing the prefactors gives

$$
N_x
=
\mathcal{N}N_p
\sqrt{4\pi\sigma_p^2}.
$$

Hence

$$
\mathcal{N}
=
\frac{N_x}{N_p\sqrt{4\pi\sigma_p^2}}.
$$

Using

$$
N_x
=
\left(
\frac{1}{2\pi\sigma_x^2}
\right)^{1/4},
$$

and

$$
N_p
=
\left(
\frac{1}{2\pi\sigma_p^2}
\right)^{1/4},
$$

we get

$$
\mathcal{N}
=
\frac{1}{2\sqrt{\pi}}
\frac{1}{\sqrt{\sigma_x\sigma_p}}.
$$

Finally, using

$$
\sigma_x\sigma_p=\frac{\hbar}{2},
$$

we obtain

$$
\mathcal{N}
=
\frac{1}{\sqrt{2\pi\hbar}}.
$$

Therefore

$$
K(x,p)
=
\frac{1}{\sqrt{2\pi\hbar}}
\exp
\left(
\frac{i}{\hbar}xp
\right).
$$

Thus the transformation from momentum representation to position representation is

$$
\psi(x)
=
\frac{1}{\sqrt{2\pi\hbar}}
\int_{-\infty}^{+\infty}
\exp
\left(
\frac{i}{\hbar}px
\right)
\phi(p)\,dp.
$$

### Fourier representation of the Dirac delta function

Using the kernel that we have just derived, we can derive an useful representation of the Dirac delta function. Starting with the completeness relation   

$$
\int_{-\infty}^{+\infty}
|p\rangle\langle p|\,dp
=
\mathbb{I}.
$$

we project onto position eigenstates to obtain  

$$
\int_{-\infty}^{+\infty}
\langle x|p\rangle
\langle p|x'\rangle\,dp
=
\langle x|x'\rangle
=
\delta(x-x').
$$

Substituting the explicit form of the kernel gives

$$
\int_{-\infty}^{+\infty}
\frac{1}{2\pi\hbar} e^{ip(x-x')/\hbar}\,dp
=
\delta(x-x').
$$  

This is the Fourier representation of the Dirac delta function. It expresses the fact that the position eigenstates $|x\rangle$ can be represented as a superposition of momentum eigenstates $|p\rangle$ with a plane wave kernel.





```python
import numpy as np
import matplotlib.pyplot as plt

from ipywidgets import interact, FloatSlider, IntSlider, Checkbox


def finite_delta(x, Pmax, hbar=1.0):
    """
    Finite-cutoff Fourier representation of the Dirac delta:

        delta_P(x) = 1/(2 pi hbar) int_{-Pmax}^{Pmax} exp(i p x / hbar) dp

    which equals

        delta_P(x) = sin(Pmax x / hbar) / (pi x)

    with the limiting value delta_P(0) = Pmax/(pi hbar).
    """

    y = np.empty_like(x)

    mask = np.abs(x) > 1e-12
    y[mask] = np.sin(Pmax * x[mask] / hbar) / (np.pi * x[mask])
    y[~mask] = Pmax / (np.pi * hbar)

    return y


def plot_delta(Pmax=50.0, x_window=2.0, hbar=1.0, show_area=True):
    x = np.linspace(-x_window, x_window, 5000)

    y = finite_delta(x, Pmax, hbar=hbar)

    plt.figure(figsize=(9, 4.5))
    plt.plot(x, y, label=r"$\delta_{P_{\max}}(x)$")
    plt.axhline(0, linewidth=0.8)

    plt.xlabel(r"$x$")
    plt.ylabel(r"$\delta_{P_{\max}}(x)$")
    plt.title(
        rf"Finite Fourier representation of $\delta(x)$, "
        rf"$P_{{\max}}={Pmax:.1f}$, $\hbar={hbar:.2f}$"
    )

    plt.legend()
    plt.tight_layout()
    plt.show()

    if show_area:
        area = np.trapezoid(y, x)
        peak = Pmax / (np.pi * hbar)
        first_zero = np.pi * hbar / Pmax

        print(f"Approximate area over plotted interval: {area:.6f}")
        print(f"Peak height delta_P(0): {peak:.6f}")
        print(f"First zero at x = ± {first_zero:.6f}")


interact(
    plot_delta,
    Pmax=FloatSlider(value=50.0, min=5.0, max=300.0, step=5.0, description="Pmax"),
    x_window=FloatSlider(value=2.0, min=0.1, max=10.0, step=0.1, description="x range"),
    hbar=FloatSlider(value=1.0, min=0.1, max=3.0, step=0.1, description="hbar"),
    show_area=Checkbox(value=True, description="show area")
)
```
