# Crash Course in Classical Mechanics

We begin our study of classical dynamics with a review of several equivalent formulations of analytical mechanics. We start with Lagrangian mechanics, then proceed to Hamiltonian mechanics, and briefly introduce the principles of Hamilton-Jacobi theory. These formulations offer complementary descriptions of the same underlying physical laws, and each has a counterpart in quantum mechanics. They have also played a fundamental role in the historical development of quantum theory and quantum dynamics.


## Hamilton's Principle and Lagrangian Mechanics

We study a classical dynamical system described by generalized coordinates
$q_1,\dots,q_n$, and let its motion be represented by a trajectory (path)
$q(t) = (q_1(t),\dots,q_n(t))$ in configuration space for a finite time interval between
$t \in [t_1,t_2]$. The dynamics of such classical system is determined by the
Lagrangian

$$
L(q,\dot q,t),
$$

and the corresponding action functional

$$
S[q] = \int_{t_1}^{t_2} L(q,\dot q,t)\,dt.
$$

Hamilton's principle states that, among all sufficiently smooth paths
connecting two fixed endpoints,

$$
q(t_1) = q^{(1)}, \qquad q(t_2) = q^{(2)},
$$

the actual path followed by the system is the one that makes the action stationary:

$$
\delta S[q] = 0.
$$

Here "stationary" means that the first-order change of the action
vanishes under arbitrary infinitesimal deformations of the path that
leave the endpoints fixed.

To make this precise, we consider a neighboring path of the form

$$
q_i(t) \longrightarrow q_i(t) + \delta q_i(t),
\qquad i=1,\dots,n,
$$

where the functions $\delta q_i(t)$ are arbitrary differentiable variations
that fullfill the endpoint conditions

$$
\delta q_i(t_1)=\delta q_i(t_2)=0.
$$

The variation commutes with the time-derivative in the sense that:

$$
\delta \dot q_i(t)=\frac{d}{dt}\delta q_i(t).
$$



The corresponding first variation of the action is given by:

$$
\delta S
=
\delta \int_{t_1}^{t_2} L(q,\dot q,t)\,dt
=
\int_{t_1}^{t_2}
\sum_{i=1}^n
\left(
\frac{\partial L}{\partial q_i}\,\delta q_i
+
\frac{\partial L}{\partial \dot q_i}\,\delta \dot q_i
\right)dt.
$$

Using

$$
\delta \dot q_i=\frac{d}{dt}\delta q_i,
$$

we obtain

$$
\delta S
=
\int_{t_1}^{t_2}
\sum_{i=1}^n
\left(
\frac{\partial L}{\partial q_i}\,\delta q_i
+
\frac{\partial L}{\partial \dot q_i}\,\frac{d}{dt}\delta q_i
\right)dt.
$$

The second term can be integrated by parts giving rise to:

$$
\int_{t_1}^{t_2}
\frac{\partial L}{\partial \dot q_i}\,\frac{d}{dt}\delta q_i\,dt
=
\left[
\frac{\partial L}{\partial \dot q_i}\,\delta q_i
\right]_{t_1}^{t_2}
-
\int_{t_1}^{t_2}
\frac{d}{dt}\left(\frac{\partial L}{\partial \dot q_i}\right)\delta q_i\,dt.
$$

Because the variations are definet such that they vanish at the endpoints, the boundary term is zero, and therefore

$$
\delta S
=
\int_{t_1}^{t_2}
\sum_{i=1}^n
\left(
\frac{\partial L}{\partial q_i}
-
\frac{d}{dt}\frac{\partial L}{\partial \dot q_i}
\right)\delta q_i\,dt.
$$

Hamilton's principle requires that $\delta S=0$ for arbitrary variations
$\delta q_i(t)$ satisfying the fixed-endpoint conditions. From this it follows that

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot q_i}
-
\frac{\partial L}{\partial q_i}
=0,
\qquad i=1,\dots,n.
$$

These are the Lagrange equations of motion, or equivalently the Euler-Lagrange equations, which form the basis of Lagrangian mechanics. When the dynamics is formulated in terms of a variational principle, the differential equations of motion arise as the necessary conditions that characterize the extremal curves of the action functional.

## Where Do We Get Lagrangians From?

A useful way to understand how Lagrangians are constructed is to begin with the simplest possible example: a free particle. From Newtonian mechanics we know that a free particle moves with constant velocity, so its trajectory is a straight line,

$$
\mathbf{r}(t)=\mathbf{r}_0+\mathbf{v}t.
$$

We may now ask: what kind of Lagrangian reproduces this motion through the Euler-Lagrange equations?

For a free particle there is no preferred position in space and no preferred time. 
It is therefore reasonable to assume that the Lagrangian of a free particle should  not depend explicitly on $\mathbf{r}$ or on $t$, and so it can depend only on the velocity:

$$
L = L(\dot{\mathbf{r}}).
$$

If, in addition, we assume that all directions in space are equivalent (space is isotropic), then no direction in space is distinguished, and the Lagrangian can depend only on the scalar combination

$$
\dot{\mathbf{r}}^{\,2}.
$$

Hence the most general form is
$$
L = f(\dot{\mathbf{r}}^{\,2}),
$$

for some function $f$.

Let us now try the simplest possible choice, namely a Lagrangian that depends linearly on the squared velocity:

$$
L = \frac{m}{2} \,\dot{\mathbf r}^{\,2},
$$

where $\frac{m}{2}$ is just a proportionality constant with a suggestive name!

 Let us now check whether this choice indeed reproduces the motion of a free particle. Since the Lagrangian does not depend explicitly on the coordinates $\mathbf r$, we have

$$
\frac{\partial L}{\partial r_i}=0,
\qquad i=1,2,3.
$$

The Euler-Lagrange equations therefore reduce to

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot r_i}=0.
$$

For the Lagrangian

$$
L=\frac{m}{2}\dot{\mathbf r}^{\,2},
$$

we find

$$
\frac{\partial L}{\partial \dot r_i}=m\dot r_i,
$$

and hence

$$
\frac{d}{dt}(m\dot r_i)=m\ddot r_i=0.
$$

Thus

$$
\ddot{\mathbf r}=0,
$$

which means that the velocity is constant and the trajectory is a straight line, exactly as expected for a free particle.

We therefore see that this simple choice of Lagrangian reproduces the correct equation of motion. Moreover, we recognize the quantity

$$
\frac{m}{2}\dot{\mathbf r}^{\,2}
$$

just as the kinetic energy of a particle of mass $m$,

$$
T=\frac{1}{2}m\mathbf v^2.
$$

Thus, for a free particle, the Lagrangian is equal to the kinetic energy:

$$
L_{\mathrm{free}}=T.
$$

We may therefore say that a free particle moves along a trajectory for which the time integral of the kinetic energy is stationary.







## What Does the Lagrangian Look Like in the General Case?

The free-particle example suggests that the Lagrangian should be built from quantities that characterize the motion of the system, namely the generalized coordinates

$$
q_1,\dots,q_n,
$$

their velocities

$$
\dot q_1,\dots,\dot q_n,
$$

and, in general, possibly the time $t$. Thus, for a system with $n$ degrees of freedom, we write

$$
L = L(q,\dot q,t).
$$

For a large class of mechanical systems, the Lagrangian takes the form

$$
L(q,\dot q,t)=T(q,\dot q,t)-V(q,t),
$$

where $T$ is the kinetic energy and $V$ is the potential energy. This is the standard form used in classical mechanics. Substituting this expression into the Euler-Lagrange equations yields the equations of motion of the system.

In Cartesian coordinates, for a single particle of mass $m$, this becomes

$$
L(\mathbf r,\dot{\mathbf r})=\frac{m}{2}\dot{\mathbf r}^{\,2}-V(\mathbf r).
$$

For several particles, the kinetic energy is the sum of the individual kinetic energies, and the potential energy may depend on all coordinates:

$$
L = \sum_{a}\frac{m_a}{2}\dot{\mathbf r}_a^{\,2} - V(\mathbf r_1,\mathbf r_2,\dots).
$$

More generally, after introducing suitable generalized coordinates, the kinetic energy need not have such a simple Cartesian form, but the Lagrangian is still often given by

$$
L=T-V.
$$



## How unique is the Lagrangian: The freedom of gauge...

We have seen that the equations of motion of a mechanical system can be derived from the condition that the action functional be stationary for a suitable choice of Lagrangian. We now show that the Lagrangian is not unique, and that there is a certain freedom in its definition.

Suppose that $L(q,\dot q,t)$ is a Lagrangian describing a given mechanical system. Let us construct a new Lagrangian by adding the total time derivative of a function that depends only on the coordinates,

$$
L'(q,\dot q,t)=L(q,\dot q,t)+\frac{d}{dt}F(q).
$$

We shall show that the equations of motion obtained from $L'$ are exactly the same as those obtained from the original Lagrangian $L$.




Let

$$
L'(q,\dot q,t)=L(q,\dot q,t)+\frac{d}{dt}F(q),
$$

where $F$ depends only on the coordinates. Since

$$
\frac{d}{dt}F(q)=\sum_{j=1}^n \frac{\partial F}{\partial q_j}\dot q_j,
$$

the new Lagrangian can be written as

$$
L'(q,\dot q,t)=L(q,\dot q,t)+\sum_{j=1}^n \frac{\partial F}{\partial q_j}\dot q_j.
$$

We now evaluate the Euler-Lagrange equation for $L'$. First,

$$
\frac{\partial L'}{\partial \dot q_i}=\frac{\partial L}{\partial \dot q_i}+\frac{\partial F}{\partial q_i},
$$

hence

$$
\frac{d}{dt}\frac{\partial L'}{\partial \dot q_i}
=\frac{d}{dt}\frac{\partial L}{\partial \dot q_i}
+\sum_{j=1}^n\frac{\partial^2 F}{\partial q_i\partial q_j}\dot q_j.
$$

Further,

$$
\frac{\partial L'}{\partial q_i}
=\frac{\partial L}{\partial q_i}
+\sum_{j=1}^n\frac{\partial^2 F}{\partial q_i\partial q_j}\dot q_j.
$$

Substituting into

$$
\frac{d}{dt}\frac{\partial L'}{\partial \dot q_i}-\frac{\partial L'}{\partial q_i}=0,
$$

the additional terms cancel and we obtain

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot q_i}-\frac{\partial L}{\partial q_i}=0.
$$

Thus the Euler-Lagrange equations derived from $L'$ are exactly the same as those derived from $L$.


The same result can be understood directly from Hamilton's principle. Let

$$
L'(q,\dot q,t)=L(q,\dot q,t)+\frac{d}{dt}F(q),
$$

with corresponding action

$$
S'[q]=\int_{t_1}^{t_2}L'\,dt
=\int_{t_1}^{t_2}L\,dt+\int_{t_1}^{t_2}\frac{d}{dt}F(q)\,dt
=S[q]+F(q(t_2))-F(q(t_1)).
$$

Taking variations,

$$
\delta S' = \delta S + \delta F(q(t_2)) - \delta F(q(t_1)).
$$

For fixed endpoints, $\delta q(t_1)=\delta q(t_2)=0$, so

$$
\delta F(q(t_1))=\delta F(q(t_2))=0,
$$

and therefore

$$
\delta S' = \delta S.
$$

Hence $\delta S'=0$ is equivalent to $\delta S=0$, and both Lagrangians yield the same equations of motion.


## Example: Harmonic Oscillator with a More Exotic Regauging

Consider the standard harmonic-oscillator Lagrangian

$$
L(q,\dot q)=\frac{m}{2}\dot q^{\,2}-\frac{k}{2}q^2,
$$

with equation of motion

$$
m\ddot q+kq=0.
$$

Add the total derivative of

$$
F(q)=\alpha\sin q,
$$

so that

$$
\frac{dF}{dt}=\alpha\cos q\,\dot q,
$$

and

$$
L'(q,\dot q)=\frac{m}{2}\dot q^{\,2}-\frac{k}{2}q^2+\alpha\cos q\,\dot q.
$$

Now,

$$
\frac{\partial L'}{\partial\dot q}=m\dot q+\alpha\cos q,
\qquad
\frac{d}{dt}\frac{\partial L'}{\partial\dot q}=m\ddot q-\alpha\sin q\,\dot q,
$$

and

$$
\frac{\partial L'}{\partial q}=-kq-\alpha\sin q\,\dot q.
$$

Therefore,

$$
\frac{d}{dt}\frac{\partial L'}{\partial\dot q}-\frac{\partial L'}{\partial q}
=\left(m\ddot q-\alpha\sin q\,\dot q\right)-\left(-kq-\alpha\sin q\,\dot q\right)
=m\ddot q+kq=0.
$$

So the regauged Lagrangian leads to exactly the same dynamics.


## Is the Classical Path a Minimum or Only Stationary?

For the harmonic oscillator the Lagrangian is given by,

$$
L(q,\dot q)=\frac{m}{2}\dot q^2-\frac{m\omega^2}{2}q^2,
$$

and the corresponding action is

$$
S[q]=\int_{t_1}^{t_2}\left(\frac{m}{2}\dot q^2-\frac{m\omega^2}{2}q^2\right)dt.
$$

Let $q_{\mathrm{cl}}$ be the classical trajectory satisfying the equation of motion $\ddot q_{\mathrm{cl}}+\omega^2 q_{\mathrm{cl}}=0$ with fixed values at $t_1$ and $t_2$. We wish to examine whether the classical path is a minimum of the action or only a stationary point. For this purpose, we consider a neighboring paths of the form

$$
q(t)=q_{\mathrm{cl}}(t)+\eta(t),\qquad \eta(t_1)=\eta(t_2)=0.
$$

Then the change in the action is given by

$$
S[q_{\mathrm{cl}}+\eta]-S[q_{\mathrm{cl}}]
=\frac{m}{2}\int_{t_1}^{t_2}\left(\dot\eta^2-\omega^2\eta^2\right)dt,
$$

because the term linear in $\eta$ vanishes on the classical path.

We now choose a particular family of variations $\eta_n$ that satisfy the boundary conditions, namely

$$
\eta_n(t)=\sin\!\left(\frac{n\pi(t-t_1)}{T}\right),\qquad T=t_2-t_1,
$$

and evaluate the change in the action for such (special) variations. We find

$$
S[q_{\mathrm{cl}}+\eta_n]-S[q_{\mathrm{cl}}]
=\frac{m}{4}T\left(\frac{n^2\pi^2}{T^2}-\omega^2\right).
$$

If $T<\pi/\omega$, this is positive (for small nonzero variations), so the classical path is a local minimum.
If $T>\pi/\omega$, it can be negative for low $n$ and positive for large $n$, so the classical path is only stationary (saddle-type), not a minimum. This illlustrates that the classical path is not always a minimum of the action, but can be a saddle point. 


## Canonical Momentum

For a system described by generalized coordinates

$$
q_1,\dots,q_n,
$$

and a Lagrangian

$$
L(q,\dot q,t),
$$

we define the **canonical momentum** conjugate to the coordinate $q_i$ by

$$
p_i=\frac{\partial L}{\partial \dot q_i}.
$$

Thus, to each generalized coordinate there corresponds a momentum obtained by differentiating the Lagrangian with respect to the corresponding generalized velocity.
Since the Lagrangian is assumed to depend on the generalized coordinates, generalized velocities, and possibly time, the canonical momentum

$$
p_i=\frac{\partial L}{\partial \dot q_i}
$$

is, in general, a function of $q$, $\dot q$, and possibly $t$:

$$
p_i=p_i(q,\dot q,t).
$$


Using the definition of the canonical momentum, the Lagrangian equation of motion can be also written as:

$$
\frac{d p_i}{dt}=\frac{\partial L}{\partial q_i}.
$$



## Examples

### Example 1: Free Particle in Cartesian Coordinates

For a free particle,

$$
L=\frac{m}{2}\dot{\mathbf r}^{\,2}
=\frac{m}{2}(\dot x^2+\dot y^2+\dot z^2),
$$

the canonical momenta are
$$

p_x=\frac{\partial L}{\partial \dot x}=m\dot x,
\qquad
p_y=\frac{\partial L}{\partial \dot y}=m\dot y,
\qquad
p_z=\frac{\partial L}{\partial \dot z}=m\dot z.
$$

In this case, the canonical momentum is just the usual linear momentum.

### Example 2: Planar Motion in Polar Coordinates

Consider a particle moving in a plane in a potential $V(r)$. Using polar coordinates $(r,\varphi)$, the Lagrangian is

$$
L=\frac{m}{2}\left(\dot r^2+r^2\dot\varphi^2\right)-V(r).
$$

The canonical momenta are

$$
p_r=\frac{\partial L}{\partial \dot r}=m\dot r,
$$

and

$$
p_\varphi=\frac{\partial L}{\partial \dot \varphi}=mr^2\dot\varphi.
$$

The quantity $p_\varphi$ is the angular momentum for the rotation around the $z$-Axis. 

### Example 3: Simple Pendulum

For a pendulum of length $\ell$, described by the angle $\theta$, the Lagrangian is

$$
L=\frac{m\ell^2}{2}\dot\theta^2-mg\ell(1-\cos\theta).
$$

The canonical momentum conjugate to $\theta$ is

$$
p_\theta=\frac{\partial L}{\partial \dot\theta}=m\ell^2\dot\theta.
$$

This is not a linear momentum, but rather the angular momentum associated with the generalized coordinate $\theta$.

### Example 4: Charged Particle in an Electromagnetic Field

Finally we look at a charged particle of charge $e$ in an electromagnetic field. Its Lagrangian is

$$
L=\frac{m}{2}\dot{\mathbf r}^{\,2}+e\,\mathbf A(\mathbf r,t)\cdot\dot{\mathbf r}-e\phi(\mathbf r,t),
$$

where $\phi$ is the scalar potential and $\mathbf A$ is the vector potential. The canonical momentum is then

$$
\mathbf p=\frac{\partial L}{\partial \dot{\mathbf r}}
= m\dot{\mathbf r}+e\,\mathbf A(\mathbf r,t).
$$

Thus the canonical momentum differs from the mechanical momentum

$$
m\dot{\mathbf r}.
$$

This is an important example showing that canonical momentum contains not only the kinematical contribution of the motion, but also information about the interaction with the field.



## Cyclic Coordinates and Conservation Laws

The concept of canonical momentum leads immediately to an important class of conservation laws. Recall that the Euler-Lagrange equations can be written in the form

$$
\frac{d}{dt}\frac{\partial L}{\partial \dot q_i}
-
\frac{\partial L}{\partial q_i}
=0,
$$

or, equivalently,

$$
\frac{d p_i}{dt}=\frac{\partial L}{\partial q_i},
$$

where

$$
p_i=\frac{\partial L}{\partial \dot q_i}
$$

is the canonical momentum conjugate to the coordinate $q_i$.

A coordinate $q_i$ is called **cyclic** (or **ignorable**) if the Lagrangian does not depend explicitly on it:

$$
\frac{\partial L}{\partial q_i}=0.
$$

In that case, the Euler-Lagrange equation reduces to

$$
\frac{d p_i}{dt}=0,
$$

and therefore

$$
p_i=\text{constant}.
$$

Thus, whenever a coordinate is cyclic, the corresponding canonical momentum is conserved.

This is one of the first clear manifestations of the relation between symmetry and conservation laws in classical mechanics: if the Lagrangian is independent of a certain coordinate (and thus invariant with respect to the change of that particular coordinate!), then the motion possesses a corresponding conserved quantity. 
In more advanced formulations, this connection between symmetry and conservation law is generalized and formalized by Noether's theorem.



## Example 1: Free Particle in Cartesian Coordinates

For a free particle,

$$
L=\frac{m}{2}\left(\dot x^2+\dot y^2+\dot z^2\right),
$$

the coordinates $x$, $y$, and $z$ do not appear explicitly in the Lagrangian. Hence they are all cyclic, and the corresponding canonical momenta are conserved:

$$
p_x=m\dot x=\text{constant}, \qquad
p_y=m\dot y=\text{constant}, \qquad
p_z=m\dot z=\text{constant}.
$$

This is simply the well known conservation of linear momentum for a free particle.



## Example 2: Central Potential in Polar Coordinates

Consider a particle moving in a plane under a central potential $V(r)$. In polar coordinates, the Lagrangian is

$$
L=\frac{m}{2}\left(\dot r^2+r^2\dot\varphi^2\right)-V(r).
$$

The angle $\varphi$ does not appear explicitly in $L$, so it is a cyclic coordinate. Therefore,

$$
p_\varphi=\frac{\partial L}{\partial \dot\varphi}=mr^2\dot\varphi=\text{constant}.
$$

This conserved quantity is the angular momentum component for rotation around a $z$-Axis.



## The Energy Function and Conservation of Energy

In the following we look at how the Lagrange-function changes as the particle moves along a classical trajectory (which solves the Lagrange's equation of motion). Let $L=L(q,\dot q,t)$ be the Lagrangian of the system. In order to see how the Lagrangian changes we differentiate it along a solution trajectory,

$$
\frac{dL}{dt}
=\sum_{i=1}^n\frac{\partial L}{\partial q_i}\dot q_i
+\sum_{i=1}^n\frac{\partial L}{\partial\dot q_i}\ddot q_i
+\frac{\partial L}{\partial t}.
$$

Using the fact that the trajectory satisfies the equation of motion,

$$
\frac{\partial L}{\partial q_i}=\frac{d}{dt}\frac{\partial L}{\partial\dot q_i},
$$

we can rewrite this as

$$
\frac{d}{dt}\left(\sum_{i=1}^n\dot q_i\frac{\partial L}{\partial\dot q_i}-L\right)
=-\frac{\partial L}{\partial t}.
$$


This expression motivates the definition of the following function that is called energy:

$$
E=\sum_{i=1}^n\dot q_i\frac{\partial L}{\partial\dot q_i}-L.
$$

With such definition we get the relation between the total time-derivative of the energy and the partial time-derivative of the Lagrangian:

$$
\frac{dE}{dt}=-\frac{\partial L}{\partial t}.
$$


In the case when the Lagrangian does not depend explicitly on time, the partial derivative vanishes and the energy is conserved along the classical trajectory!
This is another important example of the symmetry principle: For a system that is invariant under time translations (the Lagrangian does not depend explicitly on time), the energy is conserved.


For mechanical systems $L=T-V$ with velocity-independent $V$ and quadratic $T$, we can derive a more familiar form of the energy. Since $T$ is quadratic in the velocities, we have

$$
\sum_i \dot q_i\frac{\partial T}{\partial\dot q_i}=2T,
$$

hence

$$
E=\sum_i\dot q_i\frac{\partial L}{\partial\dot q_i}-L
=\sum_i\dot q_i\frac{\partial T}{\partial\dot q_i}-(T-V)
=2T-T+V=T+V.
$$

Example (harmonic oscillator):

$$
L=\frac{m}{2}\dot q^2-\frac{k}{2}q^2
\quad\Rightarrow\quad
E=\frac{m}{2}\dot q^2+\frac{k}{2}q^2,
$$

which is a conserved quantity for oscillatory trajectories.


## From Lagrangian to Hamiltonian

Lagrangian mechanics is a formulation of classical mechanics that makes clear what the underlying principle of motion is: the true trajectory (path) taken by the system is the one that makes the action stationary. 
So why do we need Hamiltonian mechanics? Hamiltonian mechanics gives a natural answer to the question how a state of a mechanical system evolves in time. It thus answers directly the question: given the state now, how does it move in time?

What is a state in classical mechanics? In Lagrangian mechanics, the state of a system at a given time is represented by a point in configuration space, which is specified by the generalized coordinates $q$. However, the equations of motion are second-order differential equations in time, so to determine the future evolution of the system we need to know not only the coordinates $q$ but also their time derivatives $\dot q$ at an some initial time. 

For mathematically inclined, the state of the mechanical system is represented by a point in the tangent bundle of configuration space, which is spanned by the coordinates and velocities $(q,\dot q)$.

In Hamiltonian mechanics, the state of the system is represented by a point in phase space, which is spanned by the coordinates and their conjugate momenta $(q,p)$. As we will see, the Hamiltonian formulation describes the time evolution of the system in terms of first-order differential equations (Hamilton's equations) in phase space.

We now switch from configuration-space variables $(q,\dot q)$ to phase-space variables $(q,p)$. Let $L(q,\dot q,t)$ be a smooth Lagrangian, and define the canonical momenta by

$$
p_i = \frac{\partial L}{\partial \dot q_i}.
$$

If the map $\dot q \mapsto p$ is locally invertible, the velocities can be written as functions $\dot q_i=\dot q_i(q,p,t)$. This means that the same dynamics can be described either by $(q,\dot q)$ or by $(q,p)$.

To see how this works, consider the total differential of $L$:

$$
dL = \sum_i \frac{\partial L}{\partial q_i}\,dq_i + \sum_i p_i\,d\dot q_i + \frac{\partial L}{\partial t}\,dt.
$$

At the same time, the Leibniz rule gives

$$
d(p_i\dot q_i)=\dot q_i\,dp_i+p_i\,d\dot q_i.
$$

I the differential of the Lagrangian we can add and subtract the term $\sum_i \dot q_i\,dp_i$, and then transfer the term $\sum_i p_i\,d\dot q_i$ to the left-hand side, giving

$$
d(\sum_i p_i\dot q_i - L) = \sum_i \dot q_i\,dp_i - \sum_i \frac{\partial L}{\partial q_i}\,dq_i - \frac{\partial L}{\partial t}\,dt.
$$

Now we define the Hamiltonian by    

$$
H(q,p,t) = \sum_i p_i\dot q_i - L(q,\dot q,t).
$$

With this definition, the left-hand side of the previous equation is just $dH$, and we have

$$
dH = \sum_i \dot q_i\,dp_i - \sum_i \frac{\partial L}{\partial q_i}\,dq_i - \frac{\partial L}{\partial t}\,dt.
$$

from which we read off the following relations:

$$
\dot q_i = \frac{\partial H}{\partial p_i},
\qquad
\frac{\partial L}{\partial q_i} = -\frac{\partial H}{\partial q_i},
\qquad
\frac{\partial L}{\partial t} = -\frac{\partial H}{\partial t}.
$$

These relations express the velocities and the derivatives of the Lagrangian in terms of the Hamiltonian.

Together with the Euler-Lagrange equations

$$
\dot p_i=\frac{\partial L}{\partial q_i},
$$

we recover Hamilton's equations

$$
\dot q_i=\frac{\partial H}{\partial p_i}, \qquad
\dot p_i=-\frac{\partial H}{\partial q_i}.
$$

In this way, the coordinate transformation that we have performed rewrites the Lagrangian dynamics as a first-order flow on phase space spanned by the coordinates and momenta. This transformation is called the Legendre transformation, and the resulting formulation of classical mechanics is called Hamiltonian mechanics.


## Classical dynamics in phase space

In Hamiltonian mechanics, the generalized coordinates and the canonical momenta are treated as independent variables. The state of a mechanical system with $f$ degrees of freedom is therefore completely specified by the $2f$ coordinates and momenta. Together, these $2f$ variables form a $2f$-dimensional phase space, in which the state of the system is represented by a single point. As the system evolves in time, this point moves along a curve in phase space, called the phase-space trajectory.


## Example: Harmonic Oscillator in Phase Space

Consider the harmonic oscillator with Lagrangian defined as 

$$
L=\frac{m}{2}\dot q^2-\frac{m \omega^2}{2}q^2.
$$

The canonical momentum is given by

$$
p=\frac{\partial L}{\partial \dot q}=m\dot q.
$$

The Hamiltonian is then obtained by performing the Legendre transformation, which gives rise to the following expression:

$$
H=p\dot q-L=\frac{p^2}{2m}+\frac{m \omega^2}{2}q^2.
$$

The Hamilton's equations of motion for the phase-space variables are

$$
\dot q=\frac{\partial H}{\partial p}=\frac{p}{m},\quad \dot p=-\frac{\partial H}{\partial q}=-m \omega^2 q.
$$



The phase-space trajectory of the harmonic oscillator is an ellipse, which is parametrized by time.

## Example: Morse-oscillator in phase space

Consider the Morse oscillator, which is a model for the vibrational motion of a diatomic molecule. Using the internuclear distance $r$ as coordinate, the Lagrangian is given by   

$$
L=\frac{\mu}{2}\dot r^2-D_e\left(1-e^{-a(r-r_e)}\right)^2,
$$  

where $\mu$ is the reduced mass, $D_e$ is the dissociation energy, $r_e$ is the equilibrium bond length, and $a$ determines the width of the potential well. The canonical momentum is

$$
p=\frac{\partial L}{\partial \dot r}=\mu\dot r,
$$

and the Hamiltonian is  

$$
H=p\dot r-L=\frac{p^2}{2\mu}+D_e\left(1-e^{-a(r-r_e)}\right)^2.
$$

The phase-space trajectories of the Morse oscillator are more complicated than those of the harmonic oscillator. We see two different regimes of the dynamics: for low energies, the trajectories are similar to those of a harmonic oscillator, while for high energies, the trajectories approach the dissociation limit.

## A more mathematical perspective on the Legendre transformation

Consider a function $f(x)$ of a single variable $x$ that is convex, meaning that its second derivative is positive: 

$$
f''(x) > 0.
$$

The convexity condition ensures that the function has a unique tangent line at each point, and that the slope of the tangent line is a monotonically increasing function of $x$. There are two equivalent ways to describe the function $f(x)$: we can specify it by giving the pair of numbers $(x, f(x))$ for each point $x$. Another equivalent way to represent the same function is by specifying the slope of the tangent line at each point together with the intercept of the tangent line $p,g(p)$. The Legendre transformation is a mathematical operation that allows us to switch between these two descriptions. 

Consider the following function of two variables:   

$$
F(x,p) = p x - f(x).
$$

That function gives the deviation of the linear function $p x$ from the original function $f(x)$. For each fixed value of $p$, we can find the value of $x$ that minimizes $F(x,p)$ by solving the equation

$$
\frac{\partial F}{\partial x} = 0 \implies p = f'(x).
$$

The convexity condition ensures that this equation has a unique solution, which we denote by $x(p)$. This gives us the point $x$ at which the linear function $p x$ is tangent to the original function $f(x)$. The value of $F$ at this point is given by

$$
g(p) = F(x(p),p) = p x(p) - f(x(p)).
$$

The function $g(p)$ is called the Legendre transform of $f(x)$. It encodes the same information as $f(x)$, but in terms of the slope $p$ of the tangent line rather than the original variable $x$. The Legendre transformation is an involution, meaning that if we apply it twice we get back the original function:

$$
f(x) = p x - g(p),
$$

where $p$ is related to $x$ by the equation

$$
p = f'(x).
$$  

In the context of classical mechanics, the Lagrangian $L(q,\dot q)$ plays the role of the function $f(x)$, where $x$ is replaced by the velocity $\dot q$. The canonical momentum $p$ is given by the derivative of the Lagrangian with respect to the velocity, and the Hamiltonian $H(q,p)$ is given by the Legendre transform of the Lagrangian. Thus, the Legendre transformation provides a systematic way to switch from the Lagrangian formulation of mechanics to the Hamiltonian formulation.  

For mathematically inclined, the Legendre transformation can be understood as a change of coordinates on the tangent bundle of configuration space, which is the natural setting for Lagrangian mechanics, to the cotangent bundle, which is the natural setting for Hamiltonian mechanics. The tangent bundle is spanned by the coordinates and velocities $(q,\dot q)$, while the cotangent bundle is spanned by the coordinates and momenta $(q,p)$. The Legendre transformation provides a canonical way to switch between these two descriptions of the same underlying physical system.


## Small Oscillations in a Molecular Potential: The Morse Potential

A realistic example of small oscillations is provided by the vibrational motion of a diatomic molecule. In this case, the interaction between the two atoms is often modeled  by the **Morse potential**

$$
V(r)=D_e\left(1-e^{-a(r-r_e)}\right)^2,
$$

where $r$ is the internuclear distance, $r_e$ is the equilibrium bond length, $D_e$ is the dissociation energy, and $a$ determines the width of the potential well.

This potential has a minimum at

$$
r=r_e,
$$

and describes not only small vibrations around equilibrium, but also the possibility of bond dissociation for large separations. It is therefore a much more realistic model than the harmonic oscillator.

To study small oscillations, we write

$$
r(t)=r_e+\xi(t),
$$

where $\xi(t)$ is a small displacement from equilibrium. The Lagrangian for the relative motion is

$$
L=\frac{\mu}{2}\dot\xi^2 - V(r_e+\xi),
$$

where

$$
\mu=\frac{m_1m_2}{m_1+m_2}
$$

is the reduced mass of the two atoms.

We now expand the Morse potential around the equilibrium point. Since

$$
e^{-a(r-r_e)}=e^{-a\xi},
$$

we use the expansion

$$
e^{-a\xi}=1-a\xi+\frac{a^2\xi^2}{2}+\cdots.
$$

Hence

$$
1-e^{-a\xi}=a\xi-\frac{a^2\xi^2}{2}+\cdots,
$$

and therefore

$$
V(r_e+\xi)
=
D_e\left(a\xi-\frac{a^2\xi^2}{2}+\cdots\right)^2.
$$

To lowest nonvanishing order, this gives
$$
V(r_e+\xi)\approx D_e a^2 \xi^2.
$$

Thus, near equilibrium, the potential becomes approximately quadratic:

$$
V(r_e+\xi)\approx \frac{k}{2}\xi^2,
$$

with

$$
k=2D_e a^2.
$$

The Lagrangian for small oscillations therefore takes the harmonic form

$$
L\approx \frac{\mu}{2}\dot\xi^2-\frac{k}{2}\xi^2,
$$

and the corresponding equation of motion is

$$
\mu \ddot\xi + k\xi = 0.
$$

Substituting the expression for $k$, we obtain

$$
\mu \ddot\xi + 2D_e a^2 \xi = 0.
$$

Hence the frequency of small vibrations is

$$
\omega=\sqrt{\frac{2D_e a^2}{\mu}}.
$$

This example shows how the harmonic oscillator emerges as a local approximation to a much more realistic molecular potential. The full Morse potential is anharmonic, but close to equilibrium its leading behavior is always harmonic. 


## Normal Modes of Polyatomic Molecules

We now extend the small-oscillation analysis from one coordinate to a polyatomic molecule with $N$ atoms.
Using Cartesian nuclear coordinates, the displacement from equilibrium is written as

$$
\Delta x_\alpha = x_\alpha - x_{0,\alpha}, \qquad \alpha=1,\dots,3N.
$$

Near equilibrium, the potential energy (eigenstate of the electronic Hamiltonian in the Born-Oppenheimer approximation!) is expanded to second order:

$$
V(\mathbf x)
\approx
V_0
+
\frac12 \sum_{\alpha\beta}
H_{\alpha\beta}\,\Delta x_\alpha\Delta x_\beta,
$$

with Hessian

$$
H_{\alpha\beta}
=
\left(\frac{\partial^2V}{\partial x_\alpha\partial x_\beta}\right)_0.
$$

The kinetic energy is

$$
T=\frac12\sum_\alpha M_\alpha\dot{\Delta x}_\alpha^2.
$$

We introduce mass-weighted coordinates

$$
q_\alpha=\sqrt{M_\alpha}\,\Delta x_\alpha,
$$

so that

$$
T=\frac12\sum_\alpha \dot q_\alpha^2,
\qquad
V=V_0+\frac12\sum_{\alpha\beta}F_{\alpha\beta}q_\alpha q_\beta,
$$

with

$$
F_{\alpha\beta}=\frac{H_{\alpha\beta}}{\sqrt{M_\alpha M_\beta}}.
$$

The equations of motion become

$$
\ddot{\mathbf q}=-\mathbf F\,\mathbf q.
$$

Using the ansatz $\mathbf q(t)=\mathbf l\,e^{i\omega t}$ gives the eigenvalue problem

$$
\mathbf F\,\mathbf l_k = \omega_k^2\,\mathbf l_k.
$$

The eigenvectors $\mathbf l_k$ define the **normal modes** of vibrations, and the eigenvalues give the squared vibrational frequencies.

For a nonlinear molecule, the $3N$ coordinates split into

- 3 translational zero modes,
- 3 rotational zero modes,
- $3N-6$ vibrational modes.

For a linear molecule, there are only 2 rotational zero modes, so the number of vibrations is $3N-5$.

### General Solution by Superposition of Eigensolutions

For each vibrational mode $k$, the two elementary solutions are

$$
\mathbf q_k^{(+)}(t)=\mathbf l_k e^{i\omega_k t},
\qquad
\mathbf q_k^{(-)}(t)=\mathbf l_k e^{-i\omega_k t}.
$$

Because the equations of motion are linear, any linear combination is again a solution. Therefore the general vibrational motion is

$$
\mathbf q(t)=\sum_{k\in\mathrm{vib}}\left(a_k e^{i\omega_k t}+b_k e^{-i\omega_k t}\right)\mathbf l_k.
$$

To determine the constants from initial values

$$
\mathbf q(0)=\mathbf q^0,
\qquad
\dot{\mathbf q}(0)=\mathbf v^0,
$$

project onto mode $j$:

$$
q_j^0\equiv \mathbf l_j\cdot\mathbf q^0 = a_j+b_j,
$$

$$
v_j^0\equiv \mathbf l_j\cdot\mathbf v^0 = i\omega_j(a_j-b_j).
$$

Hence

$$
a_j=\frac12\left(q_j^0-\frac{i}{\omega_j}v_j^0\right),
\qquad
b_j=\frac12\left(q_j^0+\frac{i}{\omega_j}v_j^0\right).
$$

Back in Cartesian coordinates,

$$
\Delta x_\alpha(t)=\frac{1}{\sqrt{M_\alpha}}
\sum_{k\in\mathrm{vib}} l_{\alpha k}
\left(a_k e^{i\omega_k t}+b_k e^{-i\omega_k t}\right).
$$

For real nuclear motion, one imposes $b_k=a_k^*$, and the physical displacement is the real part of this expression.

(If translational and rotational zero modes are included, their amplitudes are linear in time instead of oscillatory.)

### Intuitive Physical Picture of Normal Modes

A normal mode is a collective vibration pattern of the whole molecule in which all atoms move with fixed relative phase and fixed displacement ratios.

- The shape of the pattern is fixed by the eigenvector $\mathbf l_k$.
- The time dependence is harmonic with phase factor $e^{\pm i\omega_k t}$.
- If only one mode is excited, the molecule keeps exactly that spatial pattern while oscillating.
- Any complicated vibration is a superposition of these independent mode patterns.
