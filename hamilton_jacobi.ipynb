# Toward the Hamilton--Jacobi equation

We have learned that the true trajectory of a classical system is the one that makes the action functional stationary. If we now insert the true trajectory into the action functional, we obtain a **function** that depends on the initial and final endpoint coordinates and time. By fixing the initial point, this function becomes a function of the final point only. This function is denoted by $S(q,t)$, and is called the **Hamilton's principal function** in the classical mechanics literature. 
Knowing how this function changes when we slightly change the endpoint of the trajectory will lead us to the Hamilton--Jacobi (HJ) equation.

In order to derive the HJ-equation we consider now two **true classical trajectories** that start from the same initial point $(t_0,q_0)$.
The first trajectory, $q(\tau)$, ends at the point $(t,q)$. The second trajectory, $q'(\tau)$, ends at the some nearby point $(t+\Delta t,\; q+\Delta q)$. Our goal is to compute the first-order change of the action between these two trajectories. 

The action of the first trajectory is given by

$$
S(q,t)=\int_{t_0}^{t} L\bigl(q(\tau),\dot q(\tau),\tau\bigr)\,d\tau.
$$

Analogously, the action of the second trajectory is

$$
S'(q+\Delta q,t+\Delta t)=\int_{t_0}^{t+\Delta t} L\bigl(q'(\tau),\dot q'(\tau),\tau\bigr)\,d\tau.
$$

We define the difference between these two actions as

$$
\Delta S = S' - S.
$$

Thus, the total change of the action upon moving the endpoint from $(t,q)$ to $(t+\Delta t, q+\Delta q)$ is given by

$$
\Delta S
=
\int_{t_0}^{t+\Delta t} L\bigl(q'(\tau),\dot q'(\tau),\tau\bigr)\,d\tau
-
\int_{t_0}^{t} L\bigl(q(\tau),\dot q(\tau),\tau\bigr)\,d\tau.
$$


We can now first split the integration range into two subintervals, separating the first integral at $t$:

$$
\Delta S
=
\int_{t_0}^{t}
\Bigl[
L\bigl(q'(\tau),\dot q'(\tau),\tau\bigr)
-
L\bigl(q(\tau),\dot q(\tau),\tau\bigr)
\Bigr]\,d\tau
+
\int_{t}^{t+\Delta t}
L\bigl(q'(\tau),\dot q'(\tau),\tau\bigr)\,d\tau.
$$

So there are two contributions to the total change of the action:

1. a difference over the common time interval $[t_0,t]$,
2. a short extra piece from $t$ to $t+\Delta t$.

On the common interval, we can define the difference of the two trajectories by

$$
\eta(\tau)=q'(\tau)-q(\tau).
$$

Then

$$
\dot\eta(\tau)=\dot q'(\tau)-\dot q(\tau).
$$


In the next step, we expand the primed Lagrangian to first order around the original (unprimed) trajectory:

$$
L\bigl(q'(\tau),\dot q'(\tau),\tau\bigr)
\approx
L\bigl(q(\tau),\dot q(\tau),\tau\bigr)
+
\frac{\partial L}{\partial q}\,\eta(\tau)
+
\frac{\partial L}{\partial \dot q}\,\dot\eta(\tau),
$$

where all derivatives are evaluated along the original (unprimed) trajectory $q(\tau)$.

Therefore,

$$
\int_{t_0}^{t}
\Bigl[
L(q',\dot q',\tau)-L(q,\dot q,\tau)
\Bigr]\,d\tau
\approx
\int_{t_0}^{t}
\left(
\frac{\partial L}{\partial q}\,\eta
+
\frac{\partial L}{\partial \dot q}\,\dot\eta
\right)d\tau.
$$


The term containing $\dot\eta$ is transformed by partial integration:

$$
\int_{t_0}^{t}
\frac{\partial L}{\partial \dot q}\,\dot\eta\,d\tau
=
\left[
\frac{\partial L}{\partial \dot q}\,\eta
\right]_{t_0}^{t}
-
\int_{t_0}^{t}
\frac{d}{d\tau}\left(\frac{\partial L}{\partial \dot q}\right)\eta\,d\tau.
$$

Hence,

$$
\int_{t_0}^{t}
\Bigl[
L(q',\dot q',\tau)-L(q,\dot q,\tau)
\Bigr]\,d\tau
\approx
\left[
\frac{\partial L}{\partial \dot q}\,\eta
\right]_{t_0}^{t}
+
\int_{t_0}^{t}
\left(
\frac{\partial L}{\partial q}
-
\frac{d}{d\tau}\frac{\partial L}{\partial \dot q}
\right)\eta\,d\tau.
$$

Since the original path $$q(\tau)$$ is a true classical trajectory, it satisfies the Euler--Lagrange equation

$$
\frac{d}{d\tau}\frac{\partial L}{\partial \dot q}
=
\frac{\partial L}{\partial q}.
$$

Therefore the integral term vanishes:

$$
\int_{t_0}^{t}
\left(
\frac{\partial L}{\partial q}
-
\frac{d}{d\tau}\frac{\partial L}{\partial \dot q}
\right)\eta\,d\tau
=0.
$$

So only the boundary term remains:

$$
\int_{t_0}^{t}
\Bigl[
L(q',\dot q',\tau)-L(q,\dot q,\tau)
\Bigr]\,d\tau
\approx
\left[
\frac{\partial L}{\partial \dot q}\,\eta
\right]_{t_0}^{t}.
$$

Because both trajectories start at the same initial point,

$$
\eta(t_0)=0,
$$

and therefore

$$
\int_{t_0}^{t}
\Bigl[
L(q',\dot q',\tau)-L(q,\dot q,\tau)
\Bigr]\,d\tau
\approx
\left.\frac{\partial L}{\partial \dot q}\right|_{\tau=t}\eta(t).
$$



The canonical momentum along the original trajectory is

$$
p(\tau)=\frac{\partial L}{\partial \dot q}\bigl(q(\tau),\dot q(\tau),\tau\bigr).
$$

Thus at the final time $t$,

$$
\left.\frac{\partial L}{\partial \dot q}\right|_{\tau=t}=p(t).
$$

So the first contribution becomes

$$
p(t)\,\eta(t).
$$


Now consider the second piece,

$$
\int_t^{t+\Delta t} L\bigl(q'(\tau),\dot q'(\tau),\tau\bigr)\,d\tau.
$$

Since the interval is short, to first order we approximate it by

$$
\int_t^{t+\Delta t} L\bigl(q'(\tau),\dot q'(\tau),\tau\bigr)\,d\tau
\approx
L\bigl(q(t),\dot q(t),t\bigr)\,\Delta t.
$$

Thus the total change in the action is

$$
\Delta S \approx p(t)\,\eta(t)+L(t)\,\Delta t.
$$



The quantity $\eta(t)=q'(t)-q(t)$ is not yet the actual endpoint shift, because the second trajectory ends at time $t+\Delta t$, not at time $t$.

The actual endpoint condition is

$$
q'(t+\Delta t)=q+\Delta q.
$$

Expand $q'(t+\Delta t)$ around $t$:

$$
q'(t+\Delta t)
=
q'(t)+\dot q'(t)\,\Delta t+O((\Delta t)^2).
$$

Since

$$
q'(t)=q(t)+\eta(t)=q+\eta(t),
$$

we obtain

$$
q+\Delta q
=
q+\eta(t)+\dot q'(t)\,\Delta t.
$$

Therefore,

$$
\Delta q=\eta(t)+\dot q'(t)\,\Delta t.
$$

To first order, we may replace $\dot q'(t)$ by $\dot q(t)$, so

$$
\eta(t)=\Delta q-\dot q(t)\,\Delta t.
$$




Substituting into the previous expression gives

$$
\Delta S
\approx
p(t)\bigl(\Delta q-\dot q(t)\,\Delta t\bigr)+L(t)\,\Delta t.
$$

Therefore,

$$
\Delta S
\approx
p(t)\,\Delta q+\bigl(L(t)-p(t)\dot q(t)\bigr)\Delta t.
$$

Now introduce the Hamiltonian

$$
H=p\dot q-L.
$$

Hence,

$$
L-p\dot q=-H,
$$

and we finally obtain

$$
\boxed{
\Delta S\approx p(t)\,\Delta q-H(t)\,\Delta t
}
$$

to first order.

Thus by taking the limit $\Delta t\to 0$ and $\Delta q\to 0$, we can write the total change of the action as a differential of the principal function:

$$
dS = p\,dq - H\,dt.
$$

From this we can read off the partial derivatives of $S$:

$$
\frac{\partial S}{\partial q}=p,
\qquad
\frac{\partial S}{\partial t}=-H.
$$

By reading off the partial derivatives of $S$, we can write the Hamiltonian as a function of $q$ and $\partial S/\partial q$:   

$$
H=H\left(q,\frac{\partial S}{\partial q},t\right).
$$  

Substituting into the second equation gives the Hamilton--Jacobi equation:

$$
\boxed{
\frac{\partial S}{\partial t}+H\left(q,\frac{\partial S}{\partial q},t\right)=0.
}
$$

The Hamilton's principal function $S(q,t)$ is a solution of the Hamilton--Jacobi equation, and it encodes the full information about the classical dynamics of the system containing all possible trajectories. Geometrically, the classical trajectories look like rays (of light), and the Hamilton's principal function is like the wavefront of the  rays. This already suggests that $S$ coule be related to the phase of some kind of wave, which is indeed the case in quantum mechanics.
