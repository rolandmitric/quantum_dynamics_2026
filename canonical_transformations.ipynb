## Canonical transformations

The Hamilton's formulation of classical mechanics led us to the concept of the phase space, where the state of a system is described by a point in a $2N$-dimensional space with coordinates given by the generalized positions and momenta. We now raise a question about the freedom we have in choosing these coordinates. In other words, can we perform a change of coordinates in phase space that preserves the form of Hamilton's equations? The answer is yes, and such transformations are called **canonical transformations**. Consider now an arbitrary transformation of the phase space coordinates:

$$
\begin{aligned}
Q_i &= Q_i(q_1, \ldots, q_N, p_1, \ldots, p_N), \\
P_i &= P_i(q_1, \ldots, q_N, p_1, \ldots, p_N).
\end{aligned}
$$

Since the Hamilton's equations of motion are the extremals of the action functional, we can use the principle of least action to derive the conditions that the transformation must satisfy in order to be canonical. The action functional is given by:    

$$
S[q(t), p(t)] = \int_{t_1}^{t_2} \left( \sum_{i=1}^N p_i \dot{q}_i - H(q, p) \right) dt.
$$


As we know from the Lagrangian mechanics, the action functional is determined up to a total time derivative of a function of the coordinates and time. 
Genaralizing this the the Hamiltonian formalism, we know that a given transformation will certainly preserve the form of Hamilton's equations if integrand of the action functional is preserved up to a total time derivative of an arbitrary function of the phase space coordinates and time. In other words, we require that:

$$
\sum_{i=1}^N p_i \dot{q}_i - H(q, p) = \sum_{i=1}^N P_i \dot{Q}_i - K(Q, P) + \frac{dF}{dt},
$$

Specializing to one degree of freedom, we can write the above condition as: 

$$
p \dot{q} - H(q, p) = P \dot{Q} - K(Q, P) + \frac{dF}{dt},
$$

This can be written in the differential form as:

$$
p dq - H dt = P dQ - K dt + dF.
$$

Which suggests that F could be taken as a general function of the old and new coordinates, as well as time. In the particular case where $F = F(q, Q, t)$, we can write the above condition as:
$$
p dq - H dt = P dQ - K dt + \frac{\partial F}{\partial q} dq + \frac{\partial F}{\partial Q} dQ + \frac{\partial F}{\partial t} dt.
$$

Comparing the coefficients of $dq$, $dQ$ and $dt$, we obtain the following relations between the old and new coordinates:

$$
\begin{aligned}
p &= \frac{\partial F}{\partial q}, \\
P &= -\frac{\partial F}{\partial Q}, \\
K &= H + \frac{\partial F}{\partial t}.
\end{aligned}
$$

Thus by taking an arbitrary function $F(q, Q, t)$, we can generate a canonical transformation that preserves the form of Hamilton's equations!

### Example 1: 

Consider the simplest canonical transformation given by the generating function $F(q, Q) = qQ$. What kind of transformation does this generate? Using the relations above, we obtain:

$$
p = \frac{\partial F}{\partial q} = Q, \quad P = -\frac{\partial F}{\partial Q} = -q, \quad K = H.
$$  

This is a canonical transformation that exchanges the position and momentum coordinates, up to a sign! This particularly nicely illustrates the symmetry between position and momentum in classical mechanics.

### Example 2: 

Consider the canonical transformation generated by the function

$$
F(q,Q)=\frac{1}{2}m\omega q^2\frac{\cos Q}{\sin Q}.
$$

Using the relations for a type-1 generating function, we obtain

$$
\begin{aligned}
p &= \frac{\partial F}{\partial q}
   = m\omega q \frac{\cos Q}{\sin Q}, \\[4pt]
P &= -\frac{\partial F}{\partial Q}
   = \frac{1}{2}m\omega q^2\frac{1}{\sin^2 Q}, \\[4pt]
K &= H.
\end{aligned}
$$

This canonical transformation can in principle be applied to any system, but it is particularly useful for the harmonic oscillator, whose Hamiltonian is

$$
H=\frac{p^2}{2m}+\frac{1}{2}m\omega^2 q^2.
$$

From the transformation equations, we can solve for the old variables in terms of the new ones:

$$
q=\sqrt{\frac{2P}{m\omega}}\sin Q,
\qquad
p=\sqrt{2m\omega P}\cos Q.
$$

Substituting these expressions into the Hamiltonian gives

$$
K(Q,P)=H(q,p)=\omega P.
$$

The equations of motion in the new coordinates are therefore

$$
\dot Q = \frac{\partial K}{\partial P}=\omega,
\qquad
\dot P = -\frac{\partial K}{\partial Q}=0.
$$

Hence,

$$
Q(t)=\omega t+\phi,
\qquad
P(t)=P_0,
$$

where $P_0$ and $\phi$ are constants determined by the initial conditions.

Transforming back to the original coordinates, we obtain

$$
q(t)=\sqrt{\frac{2P_0}{m\omega}}\sin(\omega t+\phi),
\qquad
p(t)=\sqrt{2m\omega P_0}\cos(\omega t+\phi),
$$

which is the standard solution of the harmonic oscillator.


## Canonical transformations and the Hamilton-Jacobi equation

As we have seen in the Example 2 above, the canonical transformation generated by the function $F(q, Q)$ can be used to simplify the Hamiltonian of a system. This brings us to the question of whether we can find a canonical transformation that transforms a given Hamiltonian into a particularly simple form, such as $K=0$. If such a transformation exists, then the equations of motion in the new coordinates are trivially solved, and we can transform back to the original coordinates to obtain the solution of the original problem. We could try to find such transformation by solving some equation that the generating function must satisfy. For reasons, that will become clear later, we seek the generating function in the form $F_2(q, P, t)$, which is a function of the old coordinates and the new momenta. Incidently, such a generating function is called a type-2 generating function and can be obtained from a type-1 generating function by a Legendre transformation! The relations between the old and new coordinates in this case are given by:  

$$
\begin{aligned}
p &= \frac{\partial F_2}{\partial q}, \\
Q &= \frac{\partial F_2}{\partial P}, \\
K &= H + \frac{\partial F_2}{\partial t}.
\end{aligned}
$$

Now let us impose the condition that the new Hamiltonian is zero, $K=0$. This gives us the following equation for the generating function:  

$$
H\left(q, \frac{\partial F_2}{\partial q}\right) + \frac{\partial F_2}{\partial t} = 0.
$$

This is again the Hamilton-Jacobi equation that we have derived above from a different perspective.
Now, what kind of solutions of this equation we are looking for? Since in the new coordinates both $Q$ and $P$ are constants of motion, the solution of the Hamilton-Jacobi equation must depend on $N$ arbitrary constants, which will be identified with the new momenta $P_i$. Such a solution is called a complete solution of the Hamilton-Jacobi equation. If we can find such a solution, then we can use it to generate a canonical transformation that transforms the original Hamiltonian into zero, and thus solve the equations of motion of the original system! Let us illustrate this with an example.

### Example 3

Consider the Hamiltonian of a free particle in one dimension:

$$
H = \frac{p^2}{2m}.
$$

The Hamilton-Jacobi equation for this system is given by:

$$
\frac{1}{2m}\left(\frac{\partial S}{\partial q}\right)^2 + \frac{\partial S}{\partial t} = 0.
$$

We can solve this equation by separation of variables, by writing the solution in the form $S(q, t) = W(q) - Et$, where $E$ is a constant. Substituting this ansatz into the Hamilton-Jacobi equation gives us:

$$
\frac{1}{2m}\left(\frac{dW}{dq}\right)^2 = E.
$$

This equation can be easily solved to give us the complete solution of the Hamilton-Jacobi equation:

$$
S(q, t) = \sqrt{2mE} \, q - Et.
$$

From the solution above, we can identify the new momentum $P$ with the constant $E$, and the new coordinate $Q$ with the constant of integration that appears when we solve for $W(q)$. Hence, we have:

$$
P = E, \quad Q = \sqrt{2mE} \, q.
$$

The canonical transformation generated by the function $S(q, t)$ transforms the original Hamiltonian into

$$
K = H + \frac{\partial S}{\partial t} = 0,
$$

and the equations of motion in the new coordinates are trivially solved, since both $Q$ and $P$ are constants of motion. Transforming back to the original coordinates, we obtain the solution of the equations of motion for the free particle:

$$
q(t) = \frac{P}{\sqrt{2mP}} t + q_0,
$$

where $q_0$ is the constant of integration.


## Infinitesimal canonical transformations and symmetries

Let us now consider a special class of canonical transformations, called infinitesimal canonical transformations, which are generated by a function that is infinitesimally close to the identity transformation. Such transformation will change the phase space variables in the following way: 

$$
\begin{aligned}
Q &= q + \epsilon f(q,p), \\
P &= p + \epsilon g(q,p), \\
\end{aligned}
$$

In general, such transformation is not always canonical, but if it is, then the generating function must satisfy certain conditions. In particula, the Poisson bracket in the new coordinates must be the same as the Poisson bracket in the old coordinates, which gives us the following condition:

$$
\{Q, P\} = \{q, p\} = 1.
$$

This condition can be used to derive the following relation between the functions $f$ and $g$:

$$
\frac{\partial f}{\partial q} + \frac{\partial g}{\partial p} = 0.
$$

This relation implies that there exists a function $F(q, p)$ such that $f = \frac{\partial F}{\partial p}$ and $g = -\frac{\partial F}{\partial q}$. Hence, the infinitesimal canonical transformation can be generated by the function $F(q, p)$ as follows:

$$
\begin{aligned}
Q &= q + \epsilon \frac{\partial F}{\partial p}, \\
P &= p - \epsilon \frac{\partial F}{\partial q}.
\end{aligned}
$$

Now let us consider the effect of such an infinitesimal canonical transformation on the Hamiltonian of a system. The new Hamiltonian is given by:

$$
K = H + \epsilon \{H, F\},
$$

where $\{H, F\}$ is the Poisson bracket of $H$ and $F$. If the function $F$ is such that $\{H, F\} = 0$, then the new Hamiltonian is the same as the old Hamiltonian, $K = H$. In this case, the infinitesimal canonical transformation is a symmetry of the system, since it leaves the Hamiltonian invariant. From this it follows also that the function $F$ is a constant of motion, since its Poisson bracket with the Hamiltonian is zero. Hence, we have established a connection between symmetries and constants of motion in classical mechanics, which is a classical version of Noether's theorem!

### Example: Momentum as generator of spatial translations 

We consider the following infinitesimal canonical transformation generated by the function: 

$$ 
F(q, p) = \epsilon p, 
$$ 

What kind of transformation does this generate? 

Using the relations above, we obtain:

$$
\begin{aligned}
Q = q + \epsilon \frac{\partial F}{\partial p} = q + \epsilon, \\
P = p - \epsilon \frac{\partial F}{\partial q} = p.
\end{aligned}
$$

This is an infinitesimal canonical transformation that translates the position coordinate by an amount $\epsilon$, while leaving the momentum coordinate unchanged. Now let us consider the effect of this transformation on a general Hamiltonian $H(q, p)$. The new Hamiltonian is given by:

$$
K = H + \epsilon \{H, F\} = H + \epsilon \{H, p\} = H - \epsilon \frac{\partial H}{\partial q
}.
$$

If the Hamiltonian does not depend on the position coordinate, $\frac{\partial H}{\partial q} = 0$, then the new Hamiltonian is the same as the old Hamiltonian, $K = H$. In this case, the infinitesimal canonical transformation is a symmetry of the system, since it leaves the Hamiltonian invariant. From this it follows also that the function $F$ is a constant of motion, since its Poisson bracket with the Hamiltonian is zero. Hence, we have established that if the Hamiltonian does not depend on the position coordinate, then the momentum is a constant of motion, which is a manifestation of Noether's theorem!
